```python
import math
def probab_calc(Ua, Ub):
  if (Ua - Ub < 0 ):
    probab = 1
  else:
    probab = math.e**(Ub - Ua)  ### check units
  return probab
```

```python
import numpy as np

Poten = np.array(df_rep[0].values.tolist())
Dist= np.array(df_trial[0].values.tolist())
Poten
```

```python
# calculating probability matrix
l = len(Poten)
T = np.matrix(np.zeros((l,l)))
for i in range(l):
  for j in range(l):
    T[i,j] = probab_calc(Poten[j], Poten[i])
```

```python
import random
import bisect
```

```python
def selection(a, i, N):           # defining function to generate a new state at every time step

    use_arr = a[i-1][:N]
    #tot = np.sum(use_arr)
    #use_arr = use_arr/tot
    use_arr = np.array(use_arr.T)
    new_arr = np.cumsum(use_arr)
    new_arr = np.insert(new_arr, 0, 0)
    l = new_arr[3663]
    k = random.uniform(0,l)

    f = bisect.bisect_right(new_arr, k , 0, N+1)
    acep = use_arr[f-1]

    return f, acep
```

```python
num_steps = 1000000
N = 3663 # Number of states
os = random.randint(1, N)       # simulation
acc = 0
N_S = []
for i in range(num_steps):
    ns, acep = selection(T, os, N) 
    if (random.uniform(0,1) <= acep):
        N_S.append(ns)
        os = ns
        acc = acc + 1
print(acc)
```

```python
R_N = []
k = len(N_S)

for i in range(k):
  temp = N_S[i]
  temp2 = Dist[temp-1]
  R_N.append(temp2)
```
